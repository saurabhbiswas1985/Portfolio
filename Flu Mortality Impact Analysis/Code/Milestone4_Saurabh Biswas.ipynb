```python
#import required libraries
import pandas as pd
import numpy as np
import requests
import urllib.request, urllib.parse, urllib.error
import json
import ssl
import re
```

```python
#read State wise data csv into dataframe
df1 = pd.read_csv("Year_State_Data.csv")
df1.head(1)
```

```python
#read state full and abbriviated name csv into dataframe
df2 = pd.read_csv("state_abbreviation.csv")
df2.head(1)
```

```python
df3 = pd.merge(df1, df2, on=["State"]) # merge df1 and df2 to get the full state name
```

```python
df3
```

```python
url = 'https://datausa.io/api/data?drilldowns=State&measures=Population' # full url
headers = {}
headers['User-Agent'] = "Mozilla/5.0 (X11; Ubuntu; Linux i686; rv:48.0) Gecko/20100101 Firefox/48.0" # dummy header
```

```python
try:
    req = urllib.request.Request(url, headers = headers)
    uh = urllib.request.urlopen(req) # open the url
    data = uh.read() # READ data from url
    json_data=json.loads(data) # load json data
except urllib.error.URLError as e: # throw error for bad url open
    print(f"ERROR: {e.reason}")
```

```python
print (json.dumps(json_data, indent=2, sort_keys=True))
```

```python
State = [] # create an empty list of State
Year = [] # create an empty list of Year
Population = [] # create an empty list of population
```

```python
data =json_data["data"] # get only data portion
```

```python
for dict_data in data: # access all the elements of this list
    
  for x, y in dict_data.items(): # loop through each key/value pair in each dictionary
    # if key is State, append value into State list
    if x == 'State':
        State.append(y)
    # if key is Year, append value into Year list    
    if x == 'Year':
        Year.append(y)
    # if key is Population, append value into Population list
    if x == 'Population':
        Population.append(y)

```

```python
# get the list of tuples from State, Year, Population lists.  
# and merge them by using zip().  
list_of_tuples = list(zip(State, Year, Population))
```

```python
# Converting lists of tuples into  
# pandas Dataframe.  
df4 = pd.DataFrame(list_of_tuples, columns = ['State', 'Year', 'Population']) 
df4
```

```python
# find null value on any rows or columns
df4.isnull().sum()
```

Based on the above result there is no null value

```python
# find duplicate values in any rows
df4.duplicated().sum()
```

```python
df4['Population in Million'] = df4['Population']/1000000 # add a new column with population in million
df4
```

Based on the above result there is no Duplicate value

```python
df4.drop(['Population'], axis=1, inplace=True) # drop Population column
```

```python
df4.rename(columns = {'State':'State Name in Full'}, inplace = True) # rename a column
df4.head(5)
```

```python
df4.Year = df4.Year.astype('int64') # change data type
df4.Year
```

```python
df5 = pd.merge(df3, df4, on=["State Name in Full", "Year"]) # merge df3 and df4 to get the population
#df5=pd.concat([df3,df4],axis=0,join='outer',join_axes=["State Name in Full", "Year"],
#ignore_index=False)
```

```python
df5
```

```python
df5.to_csv('State_Details.csv', encoding='utf-8', index=False)
```
