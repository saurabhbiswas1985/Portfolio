```python
# import required libraries
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
import pandas as pd
```

We will webscarpe a website that provides short US state abbriviation and corresponding full name. Outcome of our milestone2 is a dataset that contains details on ful death, other deaths per state year wise. The state in this dataset is a two position short abbriviation. We will use the dataset from today's exercise and merge it with previous one to get full state name in the combined dataset. This full state name will be used in next week's API to fetch population for each state year wise.

```python
# read html page and extract the content

headers={'User-Agent': 'Mozilla/5.0'}  #fake header

req = Request(url="https://www.50states.com/abbreviations.htm", headers=headers) # build request with fake header

read_html = urlopen(req).read() # open url & read it
soup = BeautifulSoup(read_html) # extract html information
```

```python
find_all_table = soup.find_all('table') # extract the content of the table tag
```

```python
for link in find_all_table:  # all the class name from list of tables
    print(link.get('class'))
```

```python
abvr_table = soup.find("table", {"class": ['spaced', 'stripedRows', 'abbreviations-table', 
                                           'table', 'table-hover']}) # get the abbreviation class 
print(abvr_table.prettify()) # print the content
```

```python
data = abvr_table.tbody.find_all('tr', recursive=False)[0] # get the header data
```

```python

header1 = [] # create an empty lsit
x = data.find_all('th') # extract data with <th> tags

for th in x:
   header1.append(th.getText().strip()) # aeppend data into the list

header1 # display header1
```

```python
data = abvr_table.tbody.find_all('tr', recursive=False)[1:] # get the details data after skipping the header
```

```python
data_rows1 = [[td.get_text().strip() for td in tr.find_all('td')] for tr in data] # extract the text from <td> tag
```

```python
df1 = pd.DataFrame(data_rows1, columns=header1) # create a dataframe
df1.head() # display the dataframe
```

Rename columns so that we can merge this dataset using 2nd column name with 2nd milestone's dataset

```python
df1 = df1.rename(columns={'US State:': 'State Name in Full', 'Abbreviation:': 'State'}) # rename columns
df1.head()
```

```python
df1.shape
```

Any NaN or Null observations for any of the variables ?

```python
df1.isnull().sum()
```

Based on the above result we have 2 NaN values in both the columns. We will drop all rows having NaN in any of the variables.

```python
df1=df1.dropna()
df1.shape
```

Two records got dropped

```python
# find duplicate values in any rows
df1.duplicated().sum()
```

Based on the above result, there is no duplicate.

We will webscrape one more website. This website will provide us flu vaccine effectiveness year wise.

We will use this data to compare against the flu deatch count from milestone2's dataset.

```python
# read html page and extract the content

headers={'User-Agent': 'Mozilla/5.0'}  #fake header

req = Request(url="https://www.cdc.gov/flu/vaccines-work/effectiveness-studies.htm", headers=headers) # build request with fake header

read_html = urlopen(req).read() # open url & read it
soup = BeautifulSoup(read_html) # extract html information
```

```python
find_all_table = soup.find_all('table') # extract the content of the table tag
```

```python
for link in find_all_table:  # all the class name from list of tables
    print(link.get('class'))
```

```python
flu_table = soup.find("table", {"class": ['table', 'table-bordered', 'table-responsive', 
                                           'table-striped']}) # get the flu table class 
print(flu_table.prettify()) # print the content
```

```python
data = flu_table.tbody.find_all('tr', recursive=False)[0] # get the header data
```

```python
header1 = [] # create an empty lsit
x = data.find_all('th') # extract data with <th> tags

for th in x:
   header1.append(th.getText().strip()) # aeppend data into the list

header1 # display header1
```

```python
data = flu_table.tbody.find_all('tr', recursive=False)[1:] # get the details data after skipping the header
```

```python
data_rows1 = [[td.get_text().strip() for td in tr.find_all('td')] for tr in data] # extract the text from <td> tag
```

```python
df2 = pd.DataFrame(data_rows1, columns=header1) # create a dataframe
df2.head() # display the dataframe
```

```python
# keep only the inertesting columns
df2.drop(['Reference', 'Study Site(s)','No. of Patients‡', '95% CI'], axis=1, inplace=True)
df2.head()
```

Any NaN or Null observations for any of the variables ?

```python
df2.isnull().sum()
```

Find duplicate values in any rows

```python
df2.duplicated().sum()
```

```python
# Get the Year portion from Influenza Season†
df2[['Year', 'Discard']] = df2['Influenza Season†'].str.split("-",expand=True,)
df2.drop(['Discard'], axis=1, inplace=True)
```

```python
df2.head()
```

Now we will get the Adjusted Overall VE(%) for each year by using rolling mean with 2 windows.

```python
#df2.set_index(['Year'], inplace=True) # crate an index
```

```python
df2[['VE(%)', 'Discard']] = df2['Adjusted Overall VE (%)'].str.split("*",expand=True,)
```

```python
df2 = df2.astype({"VE(%)": int})
df2
```

```python
# Adjusted Overall VE (%) is the column on which 
# we are performing the operation 
# mean() function finds the mean over each window 
df2['VE Mean(%)']=df2['VE(%)'].rolling(2).mean() 
```

```python
# keep only the inertesting columns
df2.drop(['Adjusted Overall VE (%)', 'VE(%)', 'Discard','Influenza Season†'], axis=1, inplace=True)
df2.head()
```

```python
df2=df2.dropna() # drop NaN
df2
```

```python
df2.to_csv('Year_VE.csv', encoding='utf-8', index=False)
```
